```python
class Tiendaonline:

    def __init__(self):
        self.inventario = []
        self.ventas_totales = 0.0
    
    def agregar_producto(self, nombre, precio, cantidad):
        producto_nuevo = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
        self.inventario.append(producto_nuevo)

        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad
                return
            else:
                nuevo_producto = {'nombre': nombre,'precio': precio,'cantidad': cantidad}
    def ver_inventario(self):                   
        for producto in self.inventario:
            print(f"Nombre: {producto['nombre']},"f"Precio: ${producto['precio']},"f"Cantidad: {producto['cantidad']}")
    
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                print(f"Nombre: {producto['nombre']},"f"Precio: ${producto['precio']},"f"Cantidad: {producto['cantidad']}")
                return
        else:
            print("el producto no existe")

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad
                print("Stock actualizado")
                return
        else:
                print(" no esta en el inventario")
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                self.inventario.remove(producto)
                print("producto eliminado")
                return
        else:
                print("Producto no encontrado en el inventario")
                

```
